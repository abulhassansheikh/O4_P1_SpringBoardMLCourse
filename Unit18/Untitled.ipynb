```python
You have a data set in csv file named insurance-data.csv.
It has 6 features 'age','gender','bmi','childrens', 'smoker','region'
The data has one target variable called cost (of insurance)

Please implement a linear regression and find the accuracy
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("C:/Users/abulh/Desktop/insurance-data.csv")
data
```

```python
data = data.dropna()
data
```

```python
data.groupby("region").agg("count")
```

```python
data["smokerVal"] = 0
data["genderVal"] = 0
data["regionVal"] = 0
```

```python
data
```

```python
data["genderVal"][data["gender"]=="male"] = 0
data["genderVal"][data["gender"]=="female"] = 1 
data["smokerVal"][data["smoker"]=="yes"] = 1
data["smokerVal"][data["smoker"]=="no"] = 0
```

```python
data["regionVal"][data["region"]=="northeast"] = 0
data["regionVal"][data["region"]=="northwest"] = 1
data["regionVal"][data["region"]=="southeast"] = 2
data["regionVal"][data["region"]=="southwest"] = 3
```

```python
data[["genderVal", "regionVal", "smokerVal"]]
```

```python
len(data)
```

```python
#Smoker & region is a factor
data.groupby("cost").agg("count")
```

```python
#construct train and test set
y = data["cost"]
y
```

```python
x = data[["age", "bmi", "childrens", "genderVal", "regionVal", "smokerVal"]]
x
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
model = LinearRegression().fit(xtrain, ytrain)
prediction = model.predict(xtest)
```

```python
prediction
```

```python
accuracy = metrics.mean_squared_error(ytest, prediction)
print(accuracy)
```

```python
print(metrics.r2_score(ytest, prediction))
```
